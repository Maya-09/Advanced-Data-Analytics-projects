```python

"""This program plots a given pdf of some variable (lamda) and the corresponding gaussian/normal""" 

#importing useful libraries
from scipy.stats import poisson, norm   
import numpy as np
import matplotlib.pyplot as plt

#given that r = 0, 12, 57, 83, 131
```

```python
#define the pdf plotting function
def pdf_plot(r):
    if r == 0:
        x = np.linspace(0, 10, 1000)    #x-(lambda) values
    else:
        x = np.linspace(0, 2*r, 1000)
    p = (x**r)*np.exp(-x)/np.math.factorial(r)  #pdf (fuction) definition
    plt.plot(x, p)  #plot
```

```python
#define the gaussian plotting function
def gaussian_plot(r):
    if r == 0:
        x = np.linspace(0, 10, 1000)      #x-(lambda) values
    else:
        x = np.linspace(0, 2*r, 1000)   
    mu = r + 1                         # mean of pdf
    sigma = np.sqrt(r + 1)  #sd of pdf
    p = norm(mu, sigma)   #normal function call
    plt.plot(x, p.pdf(x))                #plot
```

```python
 #plot for r = 0 
pdf_plot(r = 0)        #pdf plot
gaussian_plot(r = 0)    #gaussian plot
plt.legend(["Pdf", "gaussian"])
plt.xlabel(r"$\lambda$")
plt.ylabel(r"p($\lambda$|r=0)")
plt.savefig("P1.png")
```
