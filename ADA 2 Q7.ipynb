```python
x = np.linspace(0, 1, 100)
def f(x, r, n):
    a = (r*(1- x)*(1-(4*(x-0.5)**2)))-((n-r)*x*(1-(4*(x-0.5)**2)))-(8*(x-0.5)*x*(1-x))
    return a
```

```python
y = np.zeros(100)
g = f(x, r = 1, n= 4)
plt.plot(x, g)
plt.plot(x, y)
plt.legend([ r"f($\theta$)", r"$f(\theta$  0.33)"])
i = 0.33
j = 0.0
plt.plot(i,j, 'ro')
plt.text(i, j, '({}, {})'.format(i, j))
plt.xlabel(r"$\theta$")
plt.ylabel(r"$f(\theta)$")
plt.grid(True)
```
